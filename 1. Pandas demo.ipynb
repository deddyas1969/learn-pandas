### Penerapan Updating Rows and Columns dari pelajaran ke 5

#### mengubah kolom header "ConvertedComp" , menjadi Salary USD

```python
df.rename(columns={'ConvertedComp': 'SalaryUSD'})
```

#### kita tambahkan inplace=True, yang artinya data di update, kalau tidak maka hanya akan membuat copy saja, 

```python
df.rename(columns={'ConvertedComp': 'SalaryUSD'}, inplace=True)
```

#### type datanya khan object, kita akan mengganti menjadi boolean Yes typenya True; dan No type nya False

#### ternyata di pandas penamaan "type data" beda dengan python, dan numpy 

#### ini link penting untuk mengenal type data dan bagaimana merubah type nya agar tepat pakai

https://pbpython.com/pandas_dtypes.html

#### kita lanjutkan, untuk mengganti Yes jadi True dan No jadi False

#### jadi intinya, kita harus cek dan ganti kalau type datanya ga sesuai

```python
df['Hobbyist'].map({'Yes' : True,'No': False})
```

### selain memakai method "map", bisa juga diganti dengan method "replace"

##### tadi di atas khan kita mengupdate isi kolom dengan
##### df.rename(columns={'ConvertedComp': 'SalaryUSD'}, inplace=True)

#### kita updatekan dengan cara lain

```python
df['Hobbyist'] = df['Hobbyist'].map({'Yes' : True,'No': False})
```
