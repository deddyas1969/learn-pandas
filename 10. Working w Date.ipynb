#### Mengetahui nilai untuk harga2 di Close, maximum untuk harga di High dsb

```python
df.resample('W').agg({'Close':'mean','High':'max','Low':'min','Volume':'sum'})
```

```python
kuncung = df.resample('W').agg({'Close':'mean','High':'max','Low':'min','Volume':'sum'})
kuncung.plot()
```

```python
high_volume = df.resample('W').agg({'High':'mean','Volume':'sum'})
high_volume.plot()
```
